```python
data_b_sorted = np.sort(data_b, order= ['yearqtr']) 
b = data_b_sorted[:np.where(data_b_sorted['yearqtr'] == '2018Q2')[0][0]]
data_b_sorted = np.append(data_b_sorted[np.where(data_b_sorted['yearqtr'] == '2018Q2')[0][0]:], b)
data_b_sorted
```

```python
# normalizamos los datos de precio 

def normalize(arr, t_min, t_max):
    norm_arr = []
    diff = t_max - t_min
    diff_arr = max(arr) - min(arr)   
    for i in arr:
        temp = (((i - min(arr))*diff)/diff_arr) + t_min
        norm_arr.append(temp)
    return norm_arr
 

price = data_b_sorted['price']
range_to_normalize = (0,1)
norm_price = normalize(price,
                                range_to_normalize[0],
                                range_to_normalize[1])
 
print("Original Array = ",price)
print("Normalized Array = ",norm_price)
```
