```python
import numpy as np
import pandas as pd
a = pd.read_csv("datos_a.csv",sep=';')
print(a.head())
print(a.dtypes)
```

```python
data_a = np.loadtxt("datos_a.csv", delimiter=";", dtype={'names':('municipio', 'fenomeno', 'total'), 'formats':(object,object,int)}, skiprows=1)
print(data_a)
print(type(data_a))
```

```python
print(data_a.shape)
print(data_a.size)
print(data_a.ndim)
print(data_a.dtype)
```

```python
print("media nacidos vivos por residencia materna:", np.mean(data_a['total'][data_a['fenomeno']=='nacidos vivos por residencia materna']))
print("media muertes fetales tardÃ\xadas por residencia materna:", np.mean(data_a['total'][data_a['fenomeno']=='muertes fetales tardÃ\xadas por residencia materna']))
print("media matrimonios por el lugar en que han fijado residencia:", np.mean(data_a['total'][data_a['fenomeno']=='matrimonios por el lugar en que han fijado residencia']))
print("media fallecidos por el lugar de residencia:", np.mean(data_a['total'][data_a['fenomeno']=='fallecidos por el lugar de residencia']))
print("media crecimiento vegetativo:", np.mean(data_a['total'][data_a['fenomeno']=='crecimiento vegetativo']))
```

```python
print("desviación nacidos vivos por residencia materna:", np.std(data_a['total'][data_a['fenomeno']=='nacidos vivos por residencia materna']))
print("desviación muertes fetales tardÃ\xadas por residencia materna:", np.std(data_a['total'][data_a['fenomeno']=='muertes fetales tardÃ\xadas por residencia materna']))
print("desviación matrimonios por el lugar en que han fijado residencia:", np.std(data_a['total'][data_a['fenomeno']=='matrimonios por el lugar en que han fijado residencia']))
print("desviación fallecidos por el lugar de residencia:", np.std(data_a['total'][data_a['fenomeno']=='fallecidos por el lugar de residencia']))
print("desviación crecimiento vegetativo:", np.std(data_a['total'][data_a['fenomeno']=='crecimiento vegetativo']))
```
